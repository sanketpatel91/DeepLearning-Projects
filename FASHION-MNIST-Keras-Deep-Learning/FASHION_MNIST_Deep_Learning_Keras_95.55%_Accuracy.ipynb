# Libraries

```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
tf.__version__
```

# Dataset

```python
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
X_train = tf.keras.utils.normalize(X_train, axis=1)
X_test = tf.keras.utils.normalize(X_test, axis=1)
```

```python
X_train.shape,X_test.shape
```

### Labels of Dataset

## Check Random images From Training Data

```python
plt.figure(figsize=(20,10))
for p in range(1,13):
    plt.subplot(3,4,p)
    img_no = random.randint(1,X_train.shape[0])
    plt.title('Image No : {0} '.format(img_no))
    plt.axis('off')
    plt.imshow(X_train[img_no])
```

## Check Random images From Test Data

```python
plt.figure(figsize=(20,10))
for p in range(1,13):
    plt.subplot(3,4,p)
    img_no = random.randint(1,X_test.shape[0])
    plt.title('Image No : {0} '.format(img_no))
    plt.axis('off')
    plt.imshow(X_test[img_no])
```

# Model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Dense(1)])
```

## Adding Layers

```python
model.add(tf.keras.layers.Flatten())

model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

## Fit the Model with epoch 200

```python
model.fit(X_train, y_train, epochs=200)
model.save("./Fashion_MNIST_DL.h5")
```

After 200 ePochs:  Epoch 200/200 => 1875/1875 [==============================] - 2s 1ms/step - loss: 0.1182 - **Accuracy: 0.9555** => **95.55%**

```python
score = model.evaluate(X_test, y_test, verbose = 0) 

print('Test loss:', round(score[0],3)) 
print('Test accuracy:', round(score[1],3))
```

## Predecting Test Data

```python
X_pred = model.predict(X_test)
```

```python
X_pred.shape
```

```python
Labels = ['T-Shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','Bag','Ankle Boot']
```

## Verify Random images of Predicated Test Data

```python
plt.figure(figsize=(30,30))
from matplotlib import rcParams
for p in range(1,21):
    plt.subplot(5,4,p)
    img_no = random.randint(1,X_test.shape[0])
    plt.title(
        "Img#: {1} & Predicted : {0}".format(
            Labels[np.where(X_pred[img_no] == np.amax(X_pred[img_no]))[0][0]],img_no),
        {'fontsize': 20})
    plt.axis('off')
    plt.imshow(X_test[img_no])

plt.savefig('FASHION_MNIST_DL_Predcting_Test_Data.png',dpi=300,facecolor='white')
```
