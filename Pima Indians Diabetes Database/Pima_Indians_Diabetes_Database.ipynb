# Libraries

```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
```

# Dataset

Location for Dataset 'diabetes.csv' : https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database?resource=download

```python
data= pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
X = data.drop('Outcome',axis=1)
y = data['Outcome']
```

```python
X.shape,y.shape
```

# Model 

## Define Model

```python
model = Sequential()
model.add(Dense(12, input_dim=8,activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, input_dim=8,activation='sigmoid'))
```

## Compile Model 

```python
#Compile
#Binary cross Entropy
#optimizer - adam / SGD
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
from keras.utils.vis_utils import plot_model
plot_model(model,show_shapes=True, show_layer_names=True)
```

##  Fit/Execute Model

```python
model.fit(X,y,epochs=10,verbose='auto',batch_size=10)
```

```python
model.fit(X,y,epochs=100,verbose='auto',batch_size=10)
```

```python
model.fit(X,y,epochs=500,verbose='auto',batch_size=10)
```

```python
model.fit(X,y,epochs=500,verbose='auto',batch_size=10)
```

```python
model.fit(X,y,epochs=500,verbose='auto',batch_size=10)
```

## Evaluate Model

```python
_,accuracy = model.evaluate(X,y)
```

```python
accuracy = accuracy * 100
print(f" Model Accuracy : {accuracy:.3f}%")
```

## Prediction 

```python
pred = model.predict(X)
```

```python
print([round(x[0]) for x in pred]) # Round to get binary output
```

```python
model.save('Diabetes_DL_ANN_Keras.h5')
```

# Prediction of  New Data with loaded Model

```python
from keras.models import load_model
loaded_model = load_model('Diabetes_DL_ANN_Keras.h5')
```

```python
Pregnancies = 1
Glucose = 126
BloodPressure = 60 
SkinThickness = 0
Insulin = 0
BMI = 23.1
DiabetesPedigreeFunction = 0.349 
Age = 56
new_data = [Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age]

new_data_1 = [Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI+10,DiabetesPedigreeFunction,Age]
```

Compare Effect of BMI with new_data vs new_data_1

```python
result = loaded_model.predict([new_data])
if result >0.5:
    print(f"Person has Diabetes with data {new_data}")
else:
    print(f"Not Having Diabetes with data {new_data}")
```

```python
result1 = loaded_model.predict([new_data_1])
if result1 >0.5:
    print(f"Person has Diabetes with data {new_data_1}")
else:
    print(f"Not Having Diabetes with data {new_data_1}")
```
